```python
uber_df=pd.read_pickle('../Resources/raw_uber_df.pkl')
uber_df.shape                  
```

```python
uber_df.head()
```

```python
uber_df['Ride_Date']= pd.to_datetime(uber_df['Pickup_date'])
```

```python
uber_df.sort_values(by=['Ride_Date'], inplace=True, ignore_index=True)
uber_df.dropna(how='any')
uber_df.head()
```

```python
# Extract year, month, day, hour, and minute. Assign all these date component to new column.

uber_df['year'] = uber_df['Ride_Date'].dt.year
uber_df['month'] = uber_df['Ride_Date'].dt.month
uber_df['day'] = uber_df['Ride_Date'].dt.day
uber_df['hour'] = uber_df['Ride_Date'].dt.hour
uber_df['Weekday_Name'] = uber_df['Ride_Date'].dt.day_name()
uber_df.head()
```

```python
import calendar
uber_df['month'] = uber_df['month'].apply(lambda x: calendar.month_abbr[x])
uber_df.head()
```

```python
uber_df.tail()
```

```python
# Weather Data for NYC 2014 from https://www.timeanddate.com/weather/usa/new-york/historic?month=1&year=2014

weather_df=pd.read_excel('../Resources/weather_data_nyc.xlsx')
weather_df
```

```python
uber_weather_df=pd.merge(uber_df, weather_df, how='inner', left_on=['year','month'], right_on=['Year','Month'])
uber_weather_df.head()
```

```python
uber_weather_df1=uber_weather_df.drop(['Dispatching_base_num','Pickup_date','Affiliated_base_num'], axis=1)
uber_weather_df1.head()
```
